```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sin(x,n):
    sinx=0
    for i in range (n):    
        sinx+=((-1)**i)*(x**(2*i+1))/math.factorial(2*i+1)
    return sinx
```

```python
angvals = np.arange(-2*np.pi,2*np.pi,0.1)
fig, ax = plt.subplots(figsize=(16,9))
actual_sin = np.sin(angvals)
ax.plot(angvals,actual_sin)
leg = ['Actual Sin']
for i in range(1,11):
    approx_sin = [sin(x,i) for x in angvals]
    ax.plot(angvals,approx_sin)
    leg.append('Approx Sin using Taylor Series - {} terms'.format(i))
ax.legend(leg)
plt.show()
```
