```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
plt.rcParams["figure.figsize"] = [16,9]
```

BAR CHART

```python
def barplot(start,end,res,width=0.4):
    year = range(start, end)
    plt.bar(year, res ,color='royalblue',width=width,label="Pass Percentage")
    plt.xlabel('YEAR')
    plt.ylabel('PASS PERCENTAGE')
    plt.title('RESULT {}-{}'.format(start,end))
    plt.grid(axis='y') 
```

```python
result = np.array(random.sample(range(50, 100), 5))
barplot(2016,2021,result)
```

PIE CHART

```python
purpose = ['Drinking','Cooking','Laundering','Cleaning','Miscellaneous'] 
usage = np.array(random.sample(range(10, 50), 5)) 
```

```python
pie(purpose,usage)
```

```python
piedonut(purpose,usage)
```
