GRADIENT DESCENT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#N_ITERATIONS AND LEARNING RATE
iterations=1000
alpha=0.01
```

```python
#DATA
m=100
x=np.linspace(0,5,m)
y=20*x+10+np.random.normal(0,2.5,np.size(x))
```

```python
#DATA POINTS PLOT
fig = plt.figure()
plt.plot(x,y,'o')
fig.suptitle('DATA POINTS')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#INITIALIZING WEIGHTS
w0 = np.random.random() #BIAS
w = np.random.rand(x[0].size)
```

```python
#COST FUNCTION(LOSS)
def J(yh, y):
    return 0.5/m*np.sum((y-yh)**2)

#FUNCTION FOR UPDATING WEIGHTS
def updateW(w0,w,yh,y,Xi,alpha):
    dw0= -alpha*np.sum(yh-y)/m
    w0=w0+dw0
    dw = -alpha*np.sum(np.dot((yh - y), Xi))/m
    w = w+dw
    return w0,w

#ROOT MEAN SQUARE ERROR
def rmse(yh, y):
    return np.sqrt(((yh - y) ** 2).mean())
```

```python
#FINDING WEIGHTS USING GD
costs = []
for i in range(iterations):
    yh=w0+w*x
    cost = J(yh,y)
    w0,w=updateW(w0,w,yh,y,x,alpha)
    costs.append(cost)
```

```python
#LOSS PLOT
fig = plt.figure()
plt.plot(costs)
fig.suptitle('LOSS PLOT')
plt.xlabel('J (w)')
plt.ylabel('# Iteration')
```

```python
#REGRESSION LINE PLOT
fig = plt.figure()
plt.plot(x,y,'o',x,yh)
fig.suptitle('REGRESSION LINE')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#RMSE
rmseGD=rmse(yh,y)
rmseGD
```

LEAST SQUARES

```python
#CALC MEAN
x_mean=x.mean()
y_mean=y.mean()
```

```python
#FIND WEIGHTS
sxy=(x*y).sum()
sxym=(x.sum()*y.sum())/m
sxx=(x*x).sum()
sxxm=(x.sum()*x.sum())/m
w=(sxy-sxym)/(sxx-sxxm)
w0=y_mean-w*x_mean
```

```python
#YHAT
yh=w0+w*x
```

```python
#PLOT
fig = plt.figure()
plt.plot(x,y,'o',x,yh)
fig.suptitle('REGRESSION LINE')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#RMSE
rmseLS=rmse(yh,y)
rmseLS
```

COMPARE RMSE

```python
print('RMSE GD = {}'.format(rmseGD))
print('RMSE LS = {}'.format(rmseLS))
```
