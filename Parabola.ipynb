```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parabola(min,max,a,b,c,ax_symmetry='y'):
    if (ax_symmetry=='y'):
        x=np.linspace(min, max, 500)
        y=a*(x**2)+b*x+c
    elif (ax_symmetry=='x'):
        y=np.linspace(min, max, 500)
        x=a*(y**2)+b*y+c
    else:
        print("WRONG INPUT")
        return
    plt.plot(x,y)
    plt.xlabel('X') 
    plt.ylabel('Y') 
    plt.title('PARABOLA Axis={} a={} b={} c={}'.format(ax_symmetry,a,b,c))
    plt.grid(True) 
```

```python
#y=x^2
parabola(-10,10,1,0,0,'y')
```

```python
fig = plt.figure(figsize=(16,10))
plt.subplot(2, 2, 1)
parabola(-10,10,1,0,0,'y')
plt.subplot(2, 2, 2)
parabola(-10,10,-1,0,0,'y')
plt.subplot(2, 2, 3)
parabola(-10,10,1,0,0,'x')
plt.subplot(2, 2, 4)
parabola(-10,10,-1,0,0,'x')
plt.show()
```

 ax_symmetry is not the axis of symmetry.
